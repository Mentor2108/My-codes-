Polynomial fit:
    Fits given x and y values of given degree n into coefficients of polynomial

```python
import numpy as np

def polynomial_fit(x, y, degree):
    # Perform polynomial fitting
    coefficients = np.polyfit(x, y, degree)

    return coefficients

# Example usage
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
degree = 1
coefficients = polynomial_fit(x, y, degree)
print(coefficients)  # prints [0., 0., 2.]
```

Calculates similarity between two poly. functions based on cosine similarity

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def calculate_similarity(coefficients1, coefficients2):
    # Convert lists to numpy arrays
    arr1 = np.array(coefficients1)
    arr2 = np.array(coefficients2)

    # Reshape arrays to row vectors
    arr1 = arr1.reshape(1, -1)
    arr2 = arr2.reshape(1, -1)

    # Calculate cosine similarity
    similarity = cosine_similarity(arr1, arr2)[0, 0]

    return similarity

# print(calculate_similarity(coefficients_Total,coefficients_four))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_polynomial(coefficients):
    # Create x values from -10 to 10
    x = np.linspace(-10, 10, 100)

    # Calculate y values using the polynomial coefficients
    y = np.polyval(coefficients, x)

    # Plot the polynomial function
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Polynomial Function')
    plt.grid(True)
    plt.show()

def plot_polynomials(coefficients1, coefficients2):
    # Create x values from -10 to 10
    x = np.linspace(-10, 10, 100)

    # Calculate y values for the first polynomial
    y1 = np.polyval(coefficients1, x)

    # Calculate y values for the second polynomial
    y2 = np.polyval(coefficients2, x)

    # Plot the polynomial functions
    plt.plot(x, y1, label='Polynomial 1')
    plt.plot(x, y2, label='Polynomial 2')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Polynomial Functions')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
# plot_polynomials(coefficients_Total,coefficients_four)
```

```python
from scipy.stats import pearsonr

def check_polynomial_similarity(poly1_coeffs, poly2_coeffs):
    correlation, _ = pearsonr(poly1_coeffs, poly2_coeffs)
    similarity = (correlation + 1) / 2  # Normalize to the range [0, 1]
    return similarity
```

```python
import random

class Particle:
    def __init__(self, dimensions, min_val, max_val):
        self.position = [int(random.uniform(min_val, max_val)) for _ in range(dimensions)]
        self.velocity = [random.uniform(min_val, max_val) for _ in range(dimensions)]
        self.best_position = self.position[:]
        self.best_fitness = float('inf')
poly_overall = polynomial_fit(x, y[2], 3)

def objective_function(X, y, poly_overall):
    if not isinstance(X, list):
        X = [X]  # Convert a single value to a list

    if not isinstance(y[0], list):
        y = [y]  # Convert a single value to a list of lists

    if len(X) > len(y[0]):
        raise ValueError("The length of X is greater than the number of y values.")

    Y = [y[0][int(X[i])] if int(X[i]) < len(y[0]) else 0.0 for i in range(len(X))]
    poly_y = polynomial_fit(X, Y, 3)
    return check_polynomial_similarity(poly_overall, poly_y)


inertia_weight = 0.5
cognitive_weight = 1.0
social_weight = 1.0


def particle_swarm_optimization(dimensions, num_particles, num_iterations, min_val, max_val, inertia_weight, cognitive_weight, social_weight):
    # Initialize particles
    particles = [Particle(dimensions, min_val, max_val) for _ in range(num_particles)]
    
    # Initialize global best
    global_best_position = [random.uniform(min_val, max_val) for _ in range(dimensions)]
    global_best_fitness = float('inf')

    # Perform optimization iterations
    for _ in range(num_iterations):
        for particle in particles:
            # Update velocity and position
            for i in range(dimensions):
                inertia = inertia_weight * particle.velocity[i]
                cognitive = cognitive_weight * random.random() * (particle.best_position[i] - particle.position[i])
                social = social_weight * random.random() * (global_best_position[i] - particle.position[i])
                particle.velocity[i] = inertia + cognitive + social
                particle.position[i] = max(min_val, min(max_val, particle.position[i] + particle.velocity[i]))  # Clamp position within bounds

            # Evaluate fitness
            polynomial_degree = dimensions - 1  # Adjust the polynomial degree based on the number of dimensions
            y_values = y[polynomial_degree]  # Select the corresponding set of y values
            fitness = objective_function(particle.position, y_values, poly_overall)

            # Update personal best
            if fitness < particle.best_fitness:
                particle.best_fitness = fitness
                particle.best_position = particle.position[:]

            # Update global best
            if fitness < global_best_fitness:
                global_best_fitness = fitness
                global_best_position = particle.position[:]

    return global_best_position, global_best_fitness



# Example usage
dimensions = 1  # Number of dimensions in the search space
num_particles = 4  # Number of particles in the swarm
num_iterations = 100  # Number of optimization iterations
min_val = 0.0  # Minimum value for each dimension
max_val = 1000.0  # Maximum value for each dimension


poly_overall = polynomial_fit(x, y[2], 3)


best_position, best_fitness = particle_swarm_optimization(dimensions, num_particles, num_iterations, min_val, max_val, inertia_weight, cognitive_weight, social_weight)
print("Best position:", best_position)
print("Best fitness:", best_fitness)
```
