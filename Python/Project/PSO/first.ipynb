Polynomial fit:
    Fits given x and y values of given degree n into coefficients of polynomial

```python
import numpy as np

def polynomial_fit(x, y, degree):
    # Perform polynomial fitting
    coefficients = np.polyfit(x, y, degree)

    return coefficients

# Example usage
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
degree = 1
coefficients = polynomial_fit(x, y, degree)
print(coefficients)  # prints [0., 0., 2.]
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_polynomial(coefficients):
    # Create x values from -10 to 10
    x = np.linspace(-10, 10, 100)

    # Calculate y values using the polynomial coefficients
    y = np.polyval(coefficients, x)

    # Plot the polynomial function
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Polynomial Function')
    plt.grid(True)
    plt.show()

def plot_polynomials(coefficients1, coefficients2):
    # Create x values from -10 to 10
    x = np.linspace(-10, 10, 100)

    # Calculate y values for the first polynomial
    y1 = np.polyval(coefficients1, x)

    # Calculate y values for the second polynomial
    y2 = np.polyval(coefficients2, x)

    # Plot the polynomial functions
    plt.plot(x, y1, label='Polynomial 1')
    plt.plot(x, y2, label='Polynomial 2')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Polynomial Functions')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_polynomials(coefficients_Total,coefficients_four)
```
